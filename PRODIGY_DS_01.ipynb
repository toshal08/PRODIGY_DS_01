```python
# Calculate population growth from 2000 to 2022
df['Growth (2000–2022)'] = df['2022 Population'] - df['2000 Population']

# Get top 10 countries by 2022 population
top_10 = df.nlargest(10, '2022 Population')

# Sort by growth
top_10_sorted = top_10.sort_values('Growth (2000–2022)', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_sorted['Country/Territory'], top_10_sorted['Growth (2000–2022)'], color='skyblue')
plt.title('Population Growth (2000–2022) in Top 10 Most Populous Countries', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Population Growth')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

# Calculate population growth from 2000 to 2022
df['Growth (2000–2022)'] = df['2022 Population'] - df['2000 Population']

# Exclude top 10 countries by 2022 population
top_10 = df.nlargest(10, '2022 Population')
others = df[~df['Country/Territory'].isin(top_10['Country/Territory'])]

# Plotting the histogram for other countries
plt.figure(figsize=(10, 6))
plt.hist(others['Growth (2000–2022)'], bins=20, color='salmon', edgecolor='black')
plt.title('Histogram of Population Growth (2000–2022)\nAll Countries Except Top 10', fontsize=14)
plt.xlabel('Population Growth')
plt.ylabel('Number of Countries')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
